```python
# Replace "no rating" string to "0" in 'product_rating' and 'overall_rating' columns otherwise dats types woun't be change.
fk['product_rating']=fk['product_rating'].str.replace("No rating available","0")
fk['overall_rating']=fk['overall_rating'].str.replace("No rating available","0")
```

```python
# Check unique values of the 'product_rating' column.
fk['product_rating'].unique
```

```python
# Check unique values of the 'overall_rating' column.
fk['overall_rating'].unique
```

```python
# Change the data types of 'product_rating' and 'overall_rating' columns.
fk=fk.astype({'product_rating':'Float64','overall_rating':'Float64'})
```

```python
# Checking data types.
fk.dtypes
```

```python
# Check unique values of the 'brand' column.
fk['brand'].unique
```

```python
# Split the 'product category tree' and replace unwanted str and save in a new column.
fk['Category'] = fk['product_category_tree'].apply(lambda x: x.split(" >> ")[0].replace('["',''))
```

```python
# Split the 'product category tree' followed by index and save in a new column.
fk['Sub_Category'] = fk['product_category_tree'].apply(lambda x: x.split(" >> ")[1:2])
```

```python
# Split the 'product category tree' followed by index and save in a new column.
fk['Subcat_6'] = fk['product_category_tree'].apply(lambda x: x.split(" >> ")[7:8])
```

```python
# Convert nan value into blank space.
fk['brand']=fk['brand'].replace(np.nan,"",regex=True)
```
